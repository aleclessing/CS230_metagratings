```python
import numpy as np
import matplotlib.pyplot as plt

from predict_tca import *
```

```python
def get_real_sr_field(model_name, scale_factor, residual=False):
    hr_eps, lr_fields, hr_fields = dataloader.MetaGratingDataLoader(return_hres=True, lr_data_filename='data/metanet_lr_data_downsamp'+str(scale_factor)+'.npy')[11]

    model = torch.load(model_name)
    sr_fields = predict_img(net=model, lr_img=lr_fields, hr_eps=hr_eps, device='cpu')

    field = sr_fields[0]
    if residual:
        field = field - hr_fields[0]

    return field
```
